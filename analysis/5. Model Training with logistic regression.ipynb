## The functions for ROC curve and confusion matrix is shown below:

```python
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(4, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(title)
    plt.show()
    
def plot_roc_curve(y_true, y_pred, label=None):
    fpr, tpr, thresholds = roc_curve(y_true, y_pred)
    plt.figure(figsize=(4, 4))
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate (Recall)')
    plt.title('ROC Curve')
    if label:
        plt.legend()
```

# 1. Logistic regression with all the rows and all the factors

```python
train_data = df[df['year'].isin([2016, 2017, 2018, 2019, 2020])]
test_data = df[df['year'] == 2021]

X_train = train_data[factors]
y_train = train_data['targeted']

X_test = test_data[factors]
y_test = test_data['targeted']

# Logistic Regression
lr = LogisticRegression(max_iter=10000)  # max_iter is set to a large value to ensure convergence
lr.fit(X_train, y_train)

# Predictions
y_train_pred = lr.predict(X_train)
y_train_pred_prob = lr.predict_proba(X_train)[:, 1]

y_test_pred = lr.predict(X_test)
y_test_pred_prob = lr.predict_proba(X_test)[:, 1]
```

```python
# Confusion Matrices
cm_train = confusion_matrix(y_train, y_train_pred)
cm_test = confusion_matrix(y_test, y_test_pred)

# Plotting
plot_confusion_matrix(cm_train, "Training Confusion Matrix")
plot_confusion_matrix(cm_test, "Test Confusion Matrix")

auc_train = roc_auc_score(y_train, y_train_pred_prob)
auc_test = roc_auc_score(y_test, y_test_pred_prob)

plot_roc_curve(y_train, y_train_pred_prob, label=f"Train ROC (AUC = {auc_train:.4f})")
plot_roc_curve(y_test, y_test_pred_prob, label=f"Test ROC (AUC = {auc_test:.4f})")
plt.legend(loc="lower right")
plt.show()
```
